Import Packages

```python
import networkx as nx
```

```python
import matplotlib.pyplot as plot
```

Create empty networkx graph

```python
G = nx.Graph()
```

Add nodes as shown in the assignment

```python
G.add_nodes_from(["Andre","Beverly","Carol","Diane","Ed","Fernando","Garth","Heather","Ike","Jane"])
```

Add edges

```python
G.add_edges_from([("Andre","Beverly"),("Andre","Diane"),("Andre","Carol"),("Andre","Fernando"),("Carol","Fernando"),("Diane","Fernando"),("Diane","Carol"),("Diane","Ed"),("Beverly","Diane"),("Beverly","Ed"),("Beverly","Garth"),("Diane","Garth"),("Ed","Garth"),("Fernando","Garth"),("Fernando","Heather"),("Garth","Heather"),("Heather","Ike"),("Ike","Jane")])
```

Show the network

```python
nx.draw(G,with_labels="True")
```
